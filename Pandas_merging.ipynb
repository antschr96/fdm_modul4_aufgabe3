Merge zweier Datens√§tze mit unterschiedlichen Optionen der Funkion: pandas.DataFrame.merge
Dokumentation der Funktion: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html


```python
import pandas as pd
```

```python
# define dataframes

# journal dataset
doi_df_journal = pd.DataFrame({
    "DOI": [
        "10.1371/journal.pcbi.1004668",
        "10.21105/joss.01035",
        "10.12688/f1000research.18866.2",
        "10.1038/s41598-019-52881-4",
        "10.1186/s12859-019-3171-0"],
    "journal": [
        "PLoS Comput Biol",
        "JOSS",
        "F1000Res",
        "Sci Rep",
        "BMC Bioinformatics"]})

# Author dataset
doi_df_first_author = pd.DataFrame({
    "DOI": [
        "10.1371/journal.pcbi.1004668",
        "10.21105/joss.01035",
        "10.21105/joss.01006",
        "10.12688/f1000research.18866.2",
        "10.1186/s12859-019-3171-0"],
    "first_author": [
        "Blischak",
        "Sparks",
        "Granger",
        "Thang",
        "Chen"]})
```

```python
doi_df_first_author
```

```python
doi_df_journal
```

```python
# Merge Dataframes - left merge
doi_df_first_author.merge(doi_df_journal, on="DOI", how="left")
# Zu jedem Autorendatensatz wird das Journal angezeigt, vorausgesetzt, die verbindende DOI ist in beiden Dataframes vorhanden, sonst wird NA angezeigt. 
```

```python
# Merge Dataframes - right merge
doi_df_first_author.merge(doi_df_journal, on="DOI", how="right")
# Enth√§lt alle Datens√§tze aus dem Journal-Datensatz, ist kein Autoren-Datensatz vorhanden, wird NA f√ºr fehlende Werte eingetragen.
```

```python
# Merge Dataframes - inner merge
doi_df_first_author.merge(doi_df_journal, on="DOI", how="inner")
# Enth√§lt nur die Datens√§tze, zu denen sowohl ein Autor als auch ein Journal-Datensatz vorhanden ist.
```

```python
# Merge Dataframes - outer merge
doi_df_first_author.merge(doi_df_journal, on="DOI", how="outer")
# Enth√§lt f√ºr jede DOI eine Zeile, jeweils fehlende Werte (Autor oder Journal) werden mit NaN ausgegeben.
```

```python
# Merge Dataframes - outer merge - changed order of data frames
doi_df_journal.merge(doi_df_first_author, on="DOI", how="outer")
# Enth√§lt f√ºr jede DOI eine Zeile, jeweils fehlende Werte (Autor oder Journal) werden mit NaN ausgegeben.
```

```python
# Merge Dataframes - cross merge (not required)
doi_df_first_author.merge(doi_df_journal, how="cross")
# jetzt noch die Variante "cross merge" ausprobieren, obwohl die keinen Sinn f√ºr diese Daten macht: sie merged jeden Datensatz aus dem einen dataframe mit jedem Datensatz des Journal Dataframes.
```
